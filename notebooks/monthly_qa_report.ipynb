# Monthly QA Automation Project
This notebook automates the monthly QA reporting process using Python and Pandas.

**Project Goals:**
- Load QA and CSAT data
- Clean and preprocess data
- Generate monthly summaries (per agent, team, and metric)
- Export results for dashboards or reports

# 1. Import Libraries

```python
# Data manipulation
import pandas as pd
import numpy as np

# Optional: for visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

# 2. Load Dataset

```python
# Load the QA/CSAT dataset from the data folder
df = pd.read_csv('../data/simulated_call_centre.csv')

# Display the first few rows to check data
df.head()
```

# 3. Explore Data

```python
# Basic info about dataset: column names, data types, missing values
df.info()

# Descriptive statistics for numeric columns
df.describe()

# Quick look at column names
df.columns
```

# 4. Data Cleaning / Preprocessing

```python
# Drop duplicates if any
df = df.drop_duplicates()

# Fill missing CSAT scores with 0 (adjust depending on your dataset)
df['csat_score'] = df['csat_score'].fillna(0)

# Convert date columns to datetime format if necessary
# df['date'] = pd.to_datetime(df['date_column'])
```

# 5. Monthly QA / CSAT Analysis

```python
# Monthly average CSAT
monthly_summary = df.groupby('month')['csat_score'].mean().reset_index()
monthly_summary.rename(columns={'csat_score': 'avg_csat'}, inplace=True)

# Monthly QA per agent
agent_summary = df.groupby(['month', 'agent_name'])['qa_score'].mean().reset_index()
```

# 6. Optional: Visualizations

```python
# Line chart of monthly average CSAT
plt.figure(figsize=(10,5))
sns.lineplot(data=monthly_summary, x='month', y='avg_csat', marker='o')
plt.title('Monthly Average CSAT')
plt.ylabel('CSAT Score')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.show()

# Heatmap for agent QA performance
agent_pivot = agent_summary.pivot(index='agent_name', columns='month', values='qa_score')
plt.figure(figsize=(12,6))
sns.heatmap(agent_pivot, annot=True, fmt=".1f", cmap='coolwarm')
plt.title('Monthly QA Scores per Agent')
plt.show()
```

# 7. Export Results

```python
# Export summaries
monthly_summary.to_csv('../data/monthly_summary.csv', index=False)
agent_summary.to_csv('../data/agent_summary.csv', index=False)
```
